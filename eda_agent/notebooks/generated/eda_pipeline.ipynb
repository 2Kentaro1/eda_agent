### データクリーニング実施

以下の手順でDataFrameをクリーニングします。

1.  **`datetime` 列のデータ型変換:**
    *   文字列型から datetime 型に変換します。
2.  **欠損値の確認と処理:**
    *   全列の欠損値の有無と割合を確認します。
    *   `payday` 列の欠損値 (`nan`) を 0 で埋めます（給料日ではないと仮定）。
    *   `precipitation` 列の " -- " を 0 に変換し、数値型に変換します。
    *   その他の欠損値（`kcal`, `remarks`, `event`）については、EDAの段階で詳しく分析し、その後のモデリングに向けて適切な処理を検討するため、今回は特定と現状維持に留めます。
3.  **データ型の調整:**
    *   数値型であるべき列が正しく数値型になっているか確認し、必要に応じて変換します。

```
import pandas as pd
import numpy as np

# DataFrameの再構築 (プレビューからの仮定)
# 実際の環境では引数としてDataFrameが渡される想定
data = {
    'datetime': ['2013-11-18', '2013-11-19', '2013-11-20', '2013-11-21', '2013-11-22'],
    'y': [90, 101, 118, 120, 130],
    'week': ['月', '火', '水', '木', '金'],
    'soldout': [0, 1, 0, 1, 1],
    'name': ['厚切りイカフライ', '手作りヒレカツ', '白身魚唐揚げ野菜あん', '若鶏ピリ辛焼', 'ビッグメンチカツ'],
    'kcal': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'remarks': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'event': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'payday': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'weather': ['快晴', '快晴', '快晴', '快晴', '快晴'],
    'precipitation': ['--', '--', '--', '--', '--'],
    'temperature': [19.8, 17.0, 15.5, 15.2, 16.1]
}
df = pd.DataFrame(data)

# 1. 'datetime' 列のデータ型変換
df['datetime'] = pd.to_datetime(df['datetime'])

# 2. 欠損値の確認と処理
print("--- 処理前の欠損値 ---")
print(df.isnull().sum())
print("\n")

# 'payday' 列の欠損値を 0 で埋める
df['payday'] = df['payday'].fillna(0)

# 'precipitation' 列の " -- " を 0 に変換し、数値型に変換
# 最初に文字列型で置換し、その後数値型に変換
df['precipitation'] = df['precipitation'].replace('--', '0').astype(float)

# 3. データ型の確認 (必要に応じて追加変換)
# 'kcal'はNaNがあるため、現時点ではfloatのままで問題なし
# 'y', 'soldout', 'temperature' は数値型として適切
# 'week', 'name', 'weather', 'remarks', 'event' はカテゴリカル/文字列として適切

print("--- 処理後の欠損値 ---")
print(df.isnull().sum())
print("\n")

print("--- 処理後のDataFrame情報 ---")
df.info()
print("\n")

print("--- クリーニング後のDataFrameの先頭5行 ---")
print(df.head())
```

### クリーニング結果の概要

*   **`datetime`**: `datetime`型に変換されました。
*   **`payday`**: 欠損値（NaN）が0で埋められ、すべて数値（0または1）になりました。
*   **`precipitation`**: " -- " という文字列が `0.0` に変換され、列のデータ型が `float64` になりました。
*   **`kcal`, `remarks`, `event`**: これらの列には引き続き欠損値が存在しますが、これは意図的に現状維持としています。EDAフェーズでその特性を詳しく調査し、その結果に基づいて後続のステップで適切な処理を検討します。

これにより、データは基本的な分析や可視化に適した形式にクリーニングされました。
